```python
import pandas as pd
import matplotlib.pyplot as plt

# Obtener las tablas de la página de Wikipedia
url = 'https://es.wikipedia.org/wiki/Anexo:Estad%C3%ADsticas_de_la_Copa_Mundial_de_F%C3%BAtbol'
tablas = pd.read_html(url)

# Seleccionar la tabla correcta (asegúrate de que esta sea la tabla correcta)
df = tablas[9]

plt.figure(figsize=(20, 4))
plt.xticks(rotation=90)
plt.bar(df['Selección'], df['Apariciones'])
plt.xlabel('Países')
plt.ylabel('Número de Apariciones en la Copa Mundial')
plt.title('Número de Participaciones de los Países en la Copa Mundial de Fútbol')
plt.show()
```

Actividad: Elaborar dataframes y comparar sus medidas de tendencia central para explicar las gráficas (Ref. Teoría cinética molecular de los gases)

boltz_maxw.PNG

```python
import pandas as pd

# Gráfica (a)

df_a = pd.DataFrame({
    "Temperatura": [100, 300, 700],
    "Velocidad media": [426.14, 519.62, 731.33],
    "Velocidad más probable": [430.64, 523.13, 734.65],
    "Velocidad rms": [459.16, 552.65, 764.17]
})

# Gráfica (b)

df_b = pd.DataFrame({
    "Gas": ["Cl2", "N2", "He"],
    "Velocidad media": [450.96, 430.64, 69.27],
    "Velocidad más probable": [453.47, 432.13, 69.58],
    "Velocidad rms": [479.37, 459.16, 71.28]
})

# Medidas de tendencia central

def medidas_tendencia_central(df):
    medidas = {
        "Media": df["Velocidad media"].mean(),
        "Moda": df["Velocidad media"].mode().iloc[0],
        "Mediana": df["Velocidad media"].median()
    }
    return medidas

# Gráfica (a)

medidas_a = medidas_tendencia_central(df_a)
print("Gráfica (a) - Medidas de tendencia central")
print(medidas_a)

# Gráfica (b)

medidas_b = medidas_tendencia_central(df_b)
print("Gráfica (b) - Medidas de tendencia central")
print(medidas_b)
```
